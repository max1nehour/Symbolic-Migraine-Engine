```python
############# RESIGN PATIENT ID ##################

import pandas as pd
import numpy as np
def add_patient_ids(csv_path, out_path=None):
    df = pd.read_csv(csv_path)

    # Add patient_id starting from 1
    df["Patient_ID"] = range(1, len(df) + 1)

    # Save if required
    if out_path:
        df.to_csv(out_path, index=False)
        print(f"Saved to {out_path}")

    return df

df = add_patient_ids("data/migraine_data.csv", "data/migraine_with_id.csv")
print(df.head())
```

```python
# Load CSV

df = pd.read_csv("data/migraine_with_id.csv")
df['Patient_ID'] = range(1, len(df) + 1)
print(df.head())

# Shuffle the dataframe
# Define split sizes

n_total = len(df)
n1 = n_total // 2
n2 = n_total - n1

df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_qwen = df_shuffled.iloc[:n1]
df_llama = df_shuffled.iloc[n1:n1+n2]

print(len(df_qwen), len(df_llama))

# Save to csv
df_qwen.to_csv("data/patients_qwen.csv", index=False)
df_llama.to_csv("data/patients_llama.csv", index=False)

print("Splitting done:")
print(f"Qwen: {len(df_qwen)}, Llama: {len(df_llama)}")
```
